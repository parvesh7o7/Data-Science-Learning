```python
#Do people who work 40+ hours per week earn more than >50K compared to those who work less than 40 hours?
#Calculate the percentage of high-earners (>50K) in each group.
#Which group has higher percentage?
```

```python
adult_db.head()
```

```python
more_hours_worker = adult_db[adult_db['hours-per-week']>=40]
```

```python
high_earner_more_hours_worker = more_hours_worker.loc[more_hours_worker['salary']==">50K"]
```

```python
high_earner_more_hours_worker
```

```python
print(f"Number of worker who works 40+ hours = {more_hours_worker.shape[0]}")
print(f"Number of worker who works 40+ hours and earns >50K = {high_earner_more_hours_worker.shape[0]}")
print(f"percentage of worker who works 40+ hours and earns >50K = {(high_earner_more_hours_worker.shape[0]/more_hours_worker.shape[0] * 100):.2f}%")
```

```python
less_hours_worker = adult_db[adult_db['hours-per-week']<40]
```

```python
high_earner_less_hours_worker = less_hours_worker.loc[less_hours_worker['salary']==">50K"]
```

```python
high_earner_less_hours_worker
```

```python
print(f"Number of worker who works 40- hours = {less_hours_worker.shape[0]}")
print(f"Number of worker who works 40- hours and earns >50K = {high_earner_less_hours_worker.shape[0]}")
print(f"percentage of worker who works 40- hours and earns >50K = {(high_earner_less_hours_worker.shape[0]/less_hours_worker.shape[0] * 100):.2f}%")
```

```python
bachelor_female_high_earner = bachelor_female.loc[bachelor_female['salary']==">50K"]
bachelor_female_high_earner_count = bachelor_female_high_earner.shape[0]
bachelor_female_high_earner_count
```
