```python
employee_data = pd.read_csv("https://raw.githubusercontent.com/rajtulluri/Employee-Promotion-Prediction/refs/heads/master/employeePromotion.csv")
```

```python
employee_data.head()
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(x='no_of_trainings', y='is_promoted', data=employee_data, hue='length_of_service')
plt.title('Employee Promotion')
```

```python
plt.figure(figsize=(12,7))
sns.lineplot(x='no_of_trainings', y='is_promoted', data=employee_data, hue='gender',palette="flare")
plt.title('Employee Promotion')
```

```python
bins = np.arange(0,1.1,0.1)
bins
```

```python
sns.distplot(employee_data['is_promoted'], bins=bins, color='red', rug=True)
plt.xticks(bins)
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
#headmaps
```

```python
flights = flights.pivot(index='month', columns='year', values='passengers')
```

```python
flights.head()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(flights, annot=True, fmt='d', linewidths=2, linecolor='k', cmap='YlGnBu')
```

```python
grid_kws = {"height_ratios":(.4, .05), "hspace": .4}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw = grid_kws)
sns.heatmap(flights, ax=ax, cbar_ax=cbar_ax, cbar_kws={"orientation":"horizontal"})
```
